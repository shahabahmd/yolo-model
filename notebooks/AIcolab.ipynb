```
import os
import random
import matplotlib.pyplot as plt
print("Train images:", len(os.listdir(f"{dataset_root}/train/images")))
print("Train labels:", len(os.listdir(f"{dataset_root}/train/labels")))
print("Val images:", len(os.listdir(f"{dataset_root}/valid/images")))
print("Val labels:", len(os.listdir(f"{dataset_root}/valid/labels")))
print("Test images:", len(os.listdir(f"{dataset_root}/test/images")))
print("Test labels:", len(os.listdir(f"{dataset_root}/test/labels")))
```

```
random_image = random.choice(os.listdir(f"{dataset_root}/train/images"))
img_path = f"{dataset_root}/train/images/{random_image}"
label_path = f"{dataset_root}/train/labels/{os.path.splitext(random_image)[0]}.txt"

print("\nRandom sample:", img_path)
print("Label file:", label_path)
```

```
img = plt.imread(img_path)
plt.imshow(img)
plt.axis("off")
plt.show()
```

```
if os.path.exists(label_path):
    with open(label_path, "r") as f:
        print("\nLabel content (YOLO format):")
        print(f.read())
else:
    print("⚠️ No label found for this image!")
```

```
label_dirs = [
    f"{dataset_root}/train/labels",
    f"{dataset_root}/valid/labels",
    f"{dataset_root}/test/labels"
]

bad_labels = []
classes_found = set()

def validate_label(line, file):
    parts = line.strip().split()
    if len(parts) != 5:
        bad_labels.append((file, "Wrong number of values", line))
        return
    try:
        cls, x, y, w, h = map(float, parts)
        cls = int(cls)
        classes_found.add(cls)
        if not (0 <= x <= 1 and 0 <= y <= 1 and 0 <= w <= 1 and 0 <= h <= 1):
            bad_labels.append((file, "Values out of [0,1] range", line))
    except ValueError:
        bad_labels.append((file, "Non-numeric values", line))
```

```
for d in label_dirs:
    if not os.path.exists(d):
        continue
    for file in os.listdir(d):
        if not file.endswith(".txt"):
            continue
        with open(os.path.join(d, file), "r") as f:
            for line in f:
                validate_label(line, file)
```

```
print("✅ Classes found in labels:", classes_found)
if bad_labels:
    print(f"⚠️ Found {len(bad_labels)} problematic labels:")
    for bl in bad_labels[:10]:  # show only first 10
        print(bl)
else:
    print("🎉 All labels look valid in YOLO format!")
```
